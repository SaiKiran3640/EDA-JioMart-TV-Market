```python
import matplotlib.pyplot as plt
import seaborn as sns

# Apply seaborn theme first
sns.set_theme(style='darkgrid')  # Apply before plotting

# Grouped summary
brand_summary = df_tv.groupby('Brand Name')['Selling Price (INR)'].agg(
    Average_Sales='mean',
    Product_Count='count'
).query('Product_Count >= 15').sort_values(by='Average_Sales', ascending=False)

# Extract values
brands = brand_summary.index.tolist()
sales = brand_summary['Average_Sales'].tolist()
counts = brand_summary['Product_Count'].tolist()

# Plotting
fig, ax1 = plt.subplots(figsize=(12, 6))

# Bar plot for Average Sales
bars = ax1.bar(brands, sales, color='skyblue')
ax1.set_ylabel('Average Price (INR)', color='blue')
ax1.tick_params(axis='y', labelcolor='blue')

# Annotate bars (above bar)
for bar, value in zip(bars, sales):
    ax1.text(bar.get_x() + bar.get_width()/2, bar.get_height() + 8000, f'{int(value):,}', 
             ha='center', va='bottom', fontsize=9, color='blue')

# Line plot for Product Count
ax2 = ax1.twinx()
ax2.plot(brands, counts, color='orange', marker='o', linewidth=2)
ax2.set_ylabel('Product Count', color='orange')
ax2.tick_params(axis='y', labelcolor='orange')

# Annotate product count points (below marker)
for i, count in enumerate(counts):
    ax2.text(i, count + 1, str(count), ha='center', va='bottom', fontsize=9, color='red')

# Final touches
plt.title('Average Price and Product Count by Brand')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Prepare data (filtered)
size_summary = df_tv.groupby('Size (Inch)')['Selling Price (INR)'].agg(
    Total_Sales='sum',
    Product_Count='count',
    Average_Price_Per_Size='mean'
).query('Product_Count > 10').sort_values(by='Total_Sales', ascending=False)

# Extract data
sizes = size_summary.index.astype(str)
total_sales = size_summary['Total_Sales']
product_counts = size_summary['Product_Count']
avg_prices = size_summary['Average_Price_Per_Size']

# Set seaborn style without grid
sns.set_style("white")

# Create plot
fig, ax1 = plt.subplots(figsize=(12, 6))

# Bar plot: Total Sales
bars = ax1.bar(sizes, total_sales, color='#1f77b4', label='Total Sales (INR)')
ax1.set_ylabel('Total Sales (INR)', color='#1f77b4')
ax1.tick_params(axis='y', labelcolor='#1f77b4')
ax1.set_axisbelow(True)

# Add value labels on bars
for bar, val in zip(bars, total_sales):
    ax1.text(bar.get_x() + bar.get_width() / 2, val + 100000, f'{int(val):,}',
             ha='center', va='bottom', fontsize=9, color='#1f77b4')

# Line plot: Product Count
ax2 = ax1.twinx()
ax2.plot(sizes, product_counts, color='#ff7f0e', marker='o', linewidth=2, label='Product Count')
ax2.set_ylabel('Product Count', color='#ff7f0e')
ax2.tick_params(axis='y', labelcolor='#ff7f0e')
for i, count in enumerate(product_counts):
    ax2.text(i, count + 5, str(count), ha='center', va='bottom', fontsize=9, color='#ff7f0e')

# Line plot: Average Price (rounded)
ax3 = ax1.twinx()
ax3.spines["right"].set_position(("axes", 1.1))
ax3.plot(sizes, avg_prices, color='#2ca02c', marker='s', linestyle='--', linewidth=2, label='Avg Price')
ax3.set_ylabel('Average Price (INR)', color='#2ca02c')
ax3.tick_params(axis='y', labelcolor="#0a130a")
for i, price in enumerate(avg_prices):
    ax3.text(i, price + 10000, f'{int(price):,}', ha='center', va='bottom', fontsize=9, color="#86221beb")

# Title and layout
plt.title('TV Size vs Total Sales, Product Count, and Avg Price')
fig.tight_layout()
fig.legend(loc='upper left', bbox_to_anchor=(0.1, 0.92), ncol=3)

# Turn off grid
ax1.grid(False)
ax2.grid(False)
ax3.grid(False)

plt.xticks(rotation=45)
plt.show()
```
